### Import Section

```python
import requests
from bs4 import BeautifulSoup
```

### Take the Date From User as an input

```python
from datetime import datetime

while True:
    Search_date = input("Enter date (YYYY-MM-DD): ")
    
    try:
        input_date = datetime.strptime(Search_date, "%Y-%m-%d") 
        current_date = datetime.today() 
        
        if input_date > current_date: 
            print("Date cannot be in the future. Please enter a valid past or present date.")
        else:
            print("Valid date format!")
            break  
    except ValueError:
        print("Invalid date format. Please enter a real date in YYYY-MM-DD format.")
```

### The main code

```python
# ex 2020-11-27
url = F"https://www.filgoal.com/matches/?date={Search_date}"
filgoalwebsite = requests.get(url).text
filgoal=BeautifulSoup(filgoalwebsite,"html") 
all=filgoal.find("div",id="match-list-viewer")
championships=all.find_all("div",class_="mc-block")
for championship in championships :
    championshiptitle=championship.find("h6").text
    print(championshiptitle)
    matches = championship.find_all("div",class_="cin_cntnr")
    print("_"*25)
    for match in matches:
        try:
            TeamA = match.find("div", class_="f").a.strong.text.strip()
            TeamB = match.find("div", class_="s").a.strong.text.strip()
            TeamAscore = match.find("div", class_="f").b.text.strip()
            TeamBscore = match.find("div", class_="s").b.text.strip()
            Time = match.find("div", class_="match-aux").find_all("span")[-1].text.strip()
            Status = match.find("span" ,class_="status over").text.strip()
            print(f"{TeamA} {TeamAscore} : {TeamBscore} {TeamB}")
            print(Time,Status)
        except:
            print("")
    print("_"*25)
```

### Lists to Construct Columns

```python
Champion_s=[] # will store the championship titles
TeamA_s=[] # will store the Teams "A"
TeamAs_score=[]# will store the score of Teams "A"
TeamBs_score=[] # will store the score of  Teams "B"
TeamB_s=[] # will store the Teams "B"
Time_s=[] # Time of each Match
Status_s=[] # Status of each Match
```

### Extracting Code

```python
for championship in championships :
    championshiptitle=championship.find("h6").text
    matches = championship.find_all("div",class_="cin_cntnr")
    for match in matches:
        try:
            Champion_s.append(championshiptitle)
            TeamA = match.find("div", class_="f").a.strong.text.strip()
            TeamB = match.find("div", class_="s").a.strong.text.strip()
            TeamAscore = match.find("div", class_="f").b.text.strip()
            TeamBscore = match.find("div", class_="s").b.text.strip()
            Time = match.find("div", class_="match-aux").find_all("span")[-1].text.strip()
            Status = match.find("span" ,class_="status over").text.strip()
            TeamA_s.append(TeamA)
            TeamAs_score.append(TeamAscore)
            TeamBs_score.append(TeamBscore)
            TeamB_s.append(TeamB)
            Time_s.append(Time)
            Status_s.append(Status)
        except:
            print("")
```

### DataFrame

```python
df=pd.DataFrame({})
```

```python
df["الفريق الاول"]=TeamA_s
df["نتيجه الاول"]=TeamAs_score
df["نتيجه الثاني"]=TeamBs_score
df["الفريق الثاني"]=TeamB_s
df["التوقيت"]=Time_s
df["الحاله"]=Status_s
df["البطوله"]=Champion_s
```
